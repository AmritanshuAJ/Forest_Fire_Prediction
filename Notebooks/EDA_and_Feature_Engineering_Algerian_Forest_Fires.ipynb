```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

#Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

##The datset is divided into sets based on region from 122nd index; so we add a new column named Region with:

###0- Bejaia Region Dataset
###1- Sidi-Bel Abbes Region Dataset

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
```

```python
dataset.info()
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
df_cleaned = dataset.dropna().reset_index(drop=True)
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.iloc[[122]]
```

```python
df_cleaned=df_cleaned.drop(122).reset_index(drop=True)
```

```python
df_cleaned.iloc[[122]]
```

```python
df_cleaned.columns
```

```python
df_cleaned.columns= df_cleaned.columns.str.strip()
```

```python
df_cleaned.columns
```

```python
df_cleaned.info()
```

```python
df_cleaned[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df_cleaned[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df_cleaned.info()
```

```python
objects=[features for features in df_cleaned.columns if df_cleaned[features].dtypes=='O']
```

```python
for i in objects:
  if i!='Classes':
    df_cleaned[i]=df_cleaned[i].astype(float)
```

```python
df_cleaned.info()
```

```python
df_cleaned.describe()
```

```python
df_cleaned.head()
```

```python
## Saving the cleaned dataset
df_cleaned.to_csv('Cleaned_Algerian_Forest_Fires_Dataset.csv',index=False)
```

# Exploratory Data Analysis(EDA)

```python
## dropping day,month,year - not required features
df_copy=df_cleaned.drop(['day','month','year'],axis=1)
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains("not fire"),0,1)
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
## Desnity plot for all the features
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
class_labels=["Fire","Not Fire"]
plt.figure(figsize=(10,5))
plt.pie(percentage,labels=class_labels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
df_copy.corr() ##correlation
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_copy.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
